````python
def create_summary(message, anthropic, max_tokens_to_sample: int = 1000):
    messages = f"""Provide a summary with about five sentences for the following article, include every important event:
        ```{message}```

        """
    resp = anthropic.completions.create(
        prompt=f"{HUMAN_PROMPT}{messages}{AI_PROMPT}",
#         prompt = messages,
        stop_sequences=[HUMAN_PROMPT],
        model="claude-v1.3",
        max_tokens_to_sample=max_tokens_to_sample,
    )
    return resp.completion.strip(' ')

def text_embedding(message):
    model = TextEmbeddingModel.from_pretrained("textembedding-gecko@001")
    embeddings = model.get_embeddings([message])
    for embedding in embeddings:
        vector = embedding.values
    return vector
````
